```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('Datasets\diabetes.xlsx')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df.head(3)
```

```python
df.columns
```

```python
unnecessary_cols=['Date', 'Time', 'Registration Number', 'Patient Name','Address', 'Contact', 'Date of Admission',
                  'Unnamed: 37','SHORT FORMS']
```

```python
df.drop(unnecessary_cols, axis=1, inplace = True)
```

```python
df.dropna(how='all', inplace= True)
```

```python

df.shape
```

```python
df.tail()
```

```python
df.replace('?', np.nan, inplace = True)
```

```python
original_label = ['m','f','Y', 'y', 'N', 'n', 'YES', 'NO','yes','no','F','M']
correct_labels = ['Male','Female','Yes', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes', 'No','Female','Male']
df.replace(original_label, correct_labels, inplace = True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
total_rows = df.shape[0]
nan_percentage = df.isnull().sum()/total_rows*100
```

```python
nan_percentage
```

```python
# Get the column names with more than 35% missing values
initial_selected_fatures = nan_percentage[nan_percentage < 15].index.tolist()
initial_selected_fatures
```

```python
diabetes1 = df[initial_selected_fatures].copy()
diabetes1['Outcome'] = 'Yes'
diabetes1.head()
```

```python
sns.boxplot(y='Weight', x='Sex',hue = 'Sex', width=0.4, data=diabetes1)
```

```python
diabetes1['Weight'].fillna(diabetes1['Weight'].median(), inplace = True)
diabetes1['Weight'].isnull().sum()
```

```python
sns.boxplot(data=diabetes1, y='Height', x='Sex',hue = 'Sex',  width=0.4)
```

```python
diabetes1['Height'].fillna(diabetes1['Height'].median(), inplace = True)
diabetes1['Height'].isnull().sum()
```

```python
sns.boxplot(data=diabetes1, y='Systolic BP', x='Sex',hue = 'Sex',  width=0.4)
```
